```python
from tqdm import tqdm 
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time

# filename = f'enlaces_changan_2024-05-14.csv'
filename = 'enlaces_volkswagen_2024-05-15.csv'  # Usar la fecha actual o asignar un valor correcto
carros_count = 0
data_list = []

# Ruta del perfil de Firefox
# profile_path = r'C:\Users\ADA_LAB01\AppData\Roaming\Mozilla\Firefox\Profiles\g45nbskz.automation'
profile_path = r'C:\Users\herna\AppData\Roaming\Mozilla\Firefox\Profiles\6mccoojc.automation'

# Configuración de opciones de Firefox
options = webdriver.FirefoxOptions()
options.add_argument('--profile')
options.add_argument(profile_path)

driver = webdriver.Firefox(options=options)

# Definir fecha de recolección antes del bloque try
fecha_recoleccion = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

try:
    df = pd.read_csv(filename)
    enlaces = df['enlace']

    for i, enlace in tqdm(enumerate(enlaces), total=len(enlaces), desc='Procesando enlaces'):
        try:
            driver.get(enlace)
            time.sleep(0.005)
            
            try:
                marca = driver.find_element(By.XPATH, '//p[small[contains(text(),"Marca")]]/span').text
            except NoSuchElementException:
                marca = "No disponible"

            try:
                modelo = driver.find_element(By.XPATH, '//p[small[contains(text(),"Modelo")]]/span').text
            except NoSuchElementException:
                modelo = "No disponible"

            try:
                nombre = driver.find_element(By.XPATH, '//*[@id="galleryHeight"]/div/div[2]/div/div/div[1]/h2').text
            except NoSuchElementException:
                nombre = "No disponible"

            try:
                precio = driver.find_element(By.XPATH, '//meta[@itemprop="price"]').get_attribute("content")
            except NoSuchElementException:
                precio = "No disponible"

            try:
                anio = driver.find_element(By.XPATH, '//*[@id="summary"]/div[2]/div[1]').text
            except NoSuchElementException:
                anio = "No disponible"

            try:
                kilometraje = driver.find_element(By.XPATH, '//meta[@itemprop="mileageFromOdometer"]').get_attribute("content")
            except NoSuchElementException:
                kilometraje = "No disponible"

            try:
                transmision = driver.find_element(By.XPATH, '//p[small[contains(text(),"Transmisión")]]/span').text
            except NoSuchElementException:
                transmision = "No disponible"

            try:
                combustible = driver.find_element(By.XPATH, '//p[small[contains(text(),"Combustible")]]/span').text
            except NoSuchElementException:
                combustible = "No disponible"

            try:
                ciudad = driver.find_element(By.XPATH, '//*[@id="summary"]/div[2]/div[2]').text
            except NoSuchElementException:
                ciudad = "No disponible"
            
            try:
                cilindraje = driver.find_element(By.XPATH, '//p[small[contains(text(),"Motor(cilindraje)")]]/span').text
            except NoSuchElementException:
                cilindraje = "No disponible"

            try:
                traccion = driver.find_element(By.XPATH, '//p[small[contains(text(),"Tracción")]]/span').text
            except NoSuchElementException:
                traccion = "No disponible"

            try:
                color = driver.find_element(By.XPATH, '//p[small[contains(text(),"Color")]]/span').text
            except NoSuchElementException:
                color = "No disponible"
            
            # Actualizar fecha de recolección dentro del bloque try
            fecha_recoleccion = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            carros_link = enlace
            
            data_list.append({
                'Marca': marca,
                'Modelo': modelo,
                'Nombre': nombre,
                'Precio': precio,
                'Anio': anio,
                'Kilometraje': kilometraje,
                'Transmision': transmision,
                'Combustible': combustible,
                'Ciudad': ciudad,
                'Cilindraje': cilindraje,
                'Traccion': traccion,
                'Color': color,
                'Fecha de recoleccion': fecha_recoleccion,
                'Link': carros_link
            })
            
            carros_count += 1
                    
        except Exception as e:
            print(f"{str(e)}")
            continue
    
    df = pd.DataFrame(data_list)
    df.to_csv(f'enlaces_volkswagen_{carros_count}.csv', index=False)
    print(f'enlaces_volkswagen_{carros_count}.csv saved')
    
except Exception as e:
    print(f"Carros: {i}. {str(e)}")
finally:
    driver.quit()
```
